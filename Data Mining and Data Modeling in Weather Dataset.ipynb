```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the weatherhistory dataset
df = pd.read_csv('weatherhistory.csv')

# Group the data by weather summary and calculate the average temperature for each category
summary_temperature = df.groupby('Summary')['Temperature (C)'].mean().reset_index()

# Sort the data by average temperature in ascending order
summary_temperature = summary_temperature.sort_values(by='Temperature (C)', ascending=True)

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(summary_temperature['Summary'], summary_temperature['Temperature (C)'])
plt.xlabel('Weather Summary')
plt.ylabel('Average Temperature (C)')
plt.title('Average Temperature by Weather Summary')
plt.xticks(rotation=90)

# Display the chart
plt.show()
```

```python


# Select the columns of interest
columns_of_interest = ['Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Pressure (millibars)']

# Create the scatter plot
plt.figure(figsize=(12, 6))
for column in columns_of_interest:
    plt.scatter(df[column], df['Temperature (C)'], label=column)

# Set labels and title
plt.xlabel('Weather Factors')
plt.ylabel('Temperature (C)')
plt.title('Relationships between Temperature and Weather Factors')

# Show legend
plt.legend()

# Display the chart
plt.show()
```

```python
import seaborn as sns
# Select the relevant columns for the analysis
columns_of_interest = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Visibility (km)', 'Pressure (millibars)']
selected_data = df[columns_of_interest]

# Compute the correlation matrix
correlation_matrix = selected_data.corr()

# Create a heat map
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap - Weather Factors')
plt.show()
```
